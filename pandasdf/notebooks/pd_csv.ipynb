```python
import pandas as pd
from pathlib import Path

csv_path = Path.cwd().parent / 'csv_files' / 'nba.csv'

df = pd.read_csv(csv_path)
# removing rows with null or NAN values
df.dropna(how="any").head()
```

```python
""" Sort_values based on column """
df.sort_values(by=["Team"])
df.sort_values(by=["Age"], ascending=False)  # sort_values by descending order.
```

```python
""" Pandas Group by """
df_team = df.groupby(['Team']).ngroups  # number of groups

df_team = df.groupby(['Team']).groups.keys() # return groups name with indexes
df_team
df_team_by_name = df.groupby('Team').get_group('Phoenix Suns').sort_values(by=["Age"])
df_team_by_name.head(5)
```
